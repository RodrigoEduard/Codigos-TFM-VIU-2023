```r
# -- 2 -- Importación de paquetes necesarios ----
library(dplyr)
library(magrittr)
library(readxl)
library(ggplot2)
```

```r
# -- 3 -- Lectura y carga del dataset ----
#Descarga del conjunto de datos:
#    https://drive.google.com/drive/folders/1nyZ_bwVZIJUMBoozzjs8pi3jO1igzb7N
conduccion <- read_excel("conduccion_110823_15_22.xlsx")
```

```r
# ZONA 1 (NORTE): XV, I, II, III, IV: 24 comunas
#   DE ARICA Y PARINACOTA(1), DE TARAPACA(3), DE ANTOFAGASTA(6), DE ATACAMA(6), DE COQUIMBO(8)
norte_filtrado <- subset(conduccion, `NOMBRE REGION` %in% c("DE ARICA Y PARINACOTA", "DE ANTOFAGASTA", "DE TARAPACA", "DE ATACAMA", "DE COQUIMBO"))
comunas_filtrado <- norte_filtrado

table_year <- table(comunas_filtrado$`NOMBRE COMUNA`, comunas_filtrado$`AÑO INGRESO`)
table_year <- as.data.frame(table_year)
table_year$Var1 <- factor(table_year$Var1, levels = unique(comunas_filtrado$`NOMBRE COMUNA`))

options(repr.plot.width=14, repr.plot.height=7.5)
table_year %>%
  ggplot(aes(x = Var2, y = Freq, group = Var1)) +
  geom_line(color = "#0b8185") +
  facet_wrap(vars(Var1), ncol = 5, scales = "free") +
  labs(x = NULL, y = NULL) +
  theme(
    legend.position = "none",
    plot.background = element_rect(fill = "white")
  )
```

```r
# ZONA 2: V, RM (CENTRO-METROPOLITANA): 24 comunas
#   DE VALPARAISO(16 comunas), METROPOLITANA DE SANTIAGO(8 comunas))
centro_metro_filtrado <- subset(conduccion, `NOMBRE REGION` %in% c("DE VALPARAISO", "METROPOLITANA DE SANTIAGO"))
comunas_filtrado <- centro_metro_filtrado

table_year <- table(comunas_filtrado$`NOMBRE COMUNA`, comunas_filtrado$`AÑO INGRESO`)
table_year <- as.data.frame(table_year)
table_year$Var1 <- factor(table_year$Var1, levels = unique(comunas_filtrado$`NOMBRE COMUNA`))

options(repr.plot.width=14, repr.plot.height=7.5)
table_year %>%
  ggplot(aes(x = Var2, y = Freq, group = Var1)) +
  geom_line(color = "#0b8185") +
  facet_wrap(vars(Var1), ncol = 5, scales = "free") +
  labs(x = NULL, y = NULL) +
  theme(
    legend.position = "none",
    plot.background = element_rect(fill = "white")
  )
```

```r
# ZONA 3: VI, VII, VII, XVI, VIII (CENTRO SUR): 47 comunas
#   DEL LIBERTADOR B. O'HIGGINS(10 comunas), DEL MAULE(11 comunas), DE ÑUBLE(6 comunas), DEL BIOBIO(20 comunas)
centro_sur_filtrado <- subset(conduccion, `NOMBRE REGION` %in% c("DEL LIBERTADOR B. O'HIGGINS", "DEL MAULE", "DE ÑUBLE", "DEL BIOBIO"))
comunas_filtrado <- centro_sur_filtrado

table_year <- table(comunas_filtrado$`NOMBRE COMUNA`, comunas_filtrado$`AÑO INGRESO`)
table_year <- as.data.frame(table_year)
table_year$Var1 <- factor(table_year$Var1, levels = unique(comunas_filtrado$`NOMBRE COMUNA`))

options(repr.plot.width=14, repr.plot.height=15)
table_year %>%
  ggplot(aes(x = Var2, y = Freq, group = Var1)) +
  geom_line(color = "#0b8185") +
  facet_wrap(vars(Var1), ncol = 5, scales = "free") +
  labs(x = NULL, y = NULL) +
  theme(
    legend.position = "none",
    plot.background = element_rect(fill = "white")
  )
```
